# Merging DataFrames

```python
# Rounding the latitude and Longitude to get a match for merging.
decimal_places = 2
city_bike['latitude'] = city_bike['latitude'].round(decimal_places)
city_bike['longitude'] = city_bike['longitude'].round(decimal_places)
fsq_df['latitude'] = fsq_df['latitude'].round(decimal_places)
fsq_df['longitude'] = fsq_df['longitude'].round(decimal_places)
yelp_df['latitude'] = yelp_df['latitude'].round(decimal_places)
yelp_df['longitude'] = yelp_df['longitude'].round(decimal_places)
```

```python
# merging the various DataFrame.

dfs_to_merge = [fsq_df,yelp_df,city_bike]
final_df = reduce(lambda left, right: pd.merge(left, right, on= ['latitude','longitude'], how='inner'), dfs_to_merge)
final_df
```

```python
def create_connection(path):
    connection = None
    try:
        connection = sqlite3.connect("my_pois.db")
        print("Connection to SQLite DB successful")
    except Error as e:
        print(f"The error '{e}' occurred")

    return connection
```

```python
 connection = create_connection("sm_app.sqlite")
```

```python
def execute_query(connection, query):
    cursor = connection.cursor()
    try:
        cursor.execute(query)
        connection.commit()
        print("Query executed successfully")
    except Error as e:
        print(f"The error '{e}' occurred")
```

```python
my_poi.to_sql('my_pois_table', con=create_connection('my_pois.db'), if_exists='replace', index=False)
```

```python
def execute_read_query(connection, query):
    cursor = connection.cursor()
    result = None
    try:
        cursor.execute(query)
        result = cursor.fetchall()  # This retrieves the results of the query
        return result
    except Error as e:
        print(f"The error '{e}' occurred")
```

### Selecting SQL Query

```python
# Create a SELECT query that fetches the first few rows
select_query = "SELECT * FROM my_pois_table LIMIT 10;"

# Pass it to the execute_read_query function
results = execute_read_query(create_connection('my_pois.db'), select_query)

# Check if you got any results back, and print them
if results:
    for row in results:
        print(row)
else:
    print("No rows found in the database.")
```
